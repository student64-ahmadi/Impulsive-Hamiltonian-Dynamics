```python
import numpy as np
import matplotlib.pyplot as plt

# Paramètres de simulation
t = np.linspace(0, 10, 1000)
gamma = 0.1
t_event = 5.0
sigma = 0.01

# Simulation (exemple simplifié)
def simulate():
    # Résultats fictifs (à remplacer par vos données)
    coherence = np.exp(-gamma * t) * (1 - 0.4 * np.exp(-(t - t_event)**2 / (2 * sigma**2)))
    population = 0.5 * (1 + np.sin(2 * np.pi * 0.32 * t))
    return coherence, population

coherence, population = simulate()

# Configuration commune des figures
plt.rcParams.update({
    'font.size': 12,
    'axes.linewidth': 1.5,
    'lines.linewidth': 2,
    'legend.frameon': False
})

# Figure 1 : Cohérence
plt.figure(figsize=(6, 4))
plt.plot(t, coherence, 'b-', label="Avec impulsion")
plt.axvline(t_event, color='gray', linestyle='--', label=r"$t_{\mathrm{event}}$")
plt.xlabel("Temps")
plt.ylabel(r"$|\rho_{01}(t)|$")
plt.xlim(0, 10)
plt.ylim(0, 1.05)
plt.legend(loc='upper right')
plt.grid(alpha=0.3)
plt.savefig("coherence.pdf", format="pdf", bbox_inches="tight")

# Figure 2 : Population
plt.figure(figsize=(6, 4))
plt.plot(t, population, 'r-', label=r"$\rho_{00}(t)$")
plt.xlabel("Temps")
plt.ylabel("Population")
plt.xlim(0, 10)
plt.ylim(0, 1.05)
plt.legend(loc='upper right')
plt.grid(alpha=0.3)
plt.savefig("population.pdf", format="pdf", bbox_inches="tight")

plt.show()
```
